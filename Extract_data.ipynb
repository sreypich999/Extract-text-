```python
from google.cloud import vision
import io
from PyPDF2 import PdfReader, PdfWriter
import tempfile

def extract_all_pages_from_pdf(pdf_path):
    """Extract text from all pages of a PDF using Google Cloud Vision (handles all pages)."""
    client = vision.ImageAnnotatorClient()
    reader = PdfReader(pdf_path)
    total_pages = len(reader.pages)
    pages_text = []

    # Process in chunks of 5 pages
    chunk_size = 5
    for start in range(0, total_pages, chunk_size):
        writer = PdfWriter()
        for i in range(start, min(start + chunk_size, total_pages)):
            writer.add_page(reader.pages[i])

        # Save chunk to a temporary PDF
        with tempfile.NamedTemporaryFile(suffix=".pdf") as tmp_pdf:
            writer.write(tmp_pdf)
            tmp_pdf.flush()
            tmp_pdf.seek(0)
            content = tmp_pdf.read()

        input_config = vision.InputConfig(content=content, mime_type="application/pdf")
        request = vision.AnnotateFileRequest(
            input_config=input_config,
            features=[vision.Feature(type_=vision.Feature.Type.DOCUMENT_TEXT_DETECTION)]
        )

        response = client.batch_annotate_files(requests=[request])

        for file_response in response.responses:
            for page_response in file_response.responses:
                text = page_response.full_text_annotation.text
                if text:
                    pages_text.append(text.strip())
                    print(f"\n===== PAGE {len(pages_text)} =====\n{text.strip()}")

    print(f"\n✅ Total pages extracted: {len(pages_text)}")
    return pages_text

# Example usage
pdf_path = "/content/បទបញ្ជាផ្ទៃក្នុងITC.pdf"
all_pages = extract_all_pages_from_pdf(pdf_path)
```

```python
import google.generativeai as genai
import os

def extract_text_from_pdf(pdf_path, api_key):
    """Extracts text from a PDF file using the Gemini 1.5 Flash model."""

    # Configure the Gemini API with your key
    try:
        genai.configure(api_key=api_key)
    except Exception as e:
        print(f"❌ Error configuring the API. Please check your API key. Details: {e}")
        return None

    # Note: The model 'gemini-2.5-flash-latest' does not exist.
    # Using 'gemini-1.5-flash-latest', the correct name for the latest Flash model.
    model = genai.GenerativeModel('gemini-2.5-flash')

    print(f"⏳ Uploading file: {pdf_path}...")
    try:
        # Upload the PDF file directly to the Gemini API
        pdf_file = genai.upload_file(path=pdf_path)
    except FileNotFoundError:
        print(f"❌ Error: The file was not found at path: {pdf_path}")
        return None
    except Exception as e:
        print(f"❌ Error uploading file: {e}")
        return None

    print("✅ File uploaded. Sending request to Gemini...")
    try:
        # Ask the model to perform OCR on the uploaded PDF
        response = model.generate_content([
            "You are an expert at data extraction. Extract all text from this PDF document. Preserve the original layout, paragraphs, and line breaks as closely as possible.",
            pdf_file
        ])

        extracted_text = response.text
        print("\n--- Extracted Text ---")
        print(extracted_text)

        # Clean up the uploaded file after use
        genai.delete_file(pdf_file.name)
        print("\n--- Uploaded file cleaned up. ---")

        return extracted_text

    except Exception as e:
        print(f"❌ An error occurred during the API call: {e}")
        # Clean up the file even if the API call fails
        genai.delete_file(pdf_file.name)
        return None

# --- How to Run ---

# 1. ⚠️ REVOKE your old key and get a new one from Google AI Studio.
# 2. Replace "YOUR_API_KEY" below with your new, private key.
api_key = ""

# 3. Set the path to your PDF file.
pdf_path = "/content/បទបញ្ជាផ្ទៃក្នុងITC.pdf"

# 4. Call the function
if api_key == "YOUR_API_KEY":
    print("❌ Please replace 'YOUR_API_KEY' with your actual Gemini API key.")
else:
    extract_text_from_pdf(pdf_path, api_key)
```

```python
# ===== STEP 2: Import libraries =====
from pdf2image import convert_from_path
import pytesseract

# ===== STEP 3: Define function to extract Khmer + English text =====
def extract_text_from_pdf(pdf_path, dpi=300):
    """
    Extracts all pages from a PDF supporting Khmer and English.
    Returns a single string with page separators.
    """
    images = convert_from_path(pdf_path, dpi=dpi)
    all_text = ""

    # Khmer + English OCR languages
    lang = 'khm+eng'
    tessdata_dir_config = '--tessdata-dir "/content/tessdata"'

    for i, image in enumerate(images, start=1):
        try:
            text = pytesseract.image_to_string(image, lang=lang, config=tessdata_dir_config)
            formatted_text = "\n".join([line.expandtabs(4) for line in text.splitlines()])
            all_text += f"\n\n===== PAGE {i} =====\n"
            all_text += formatted_text
        except Exception as e:
            print(f"Error during OCR on page {i}: {e}")

    return all_text

# ===== STEP 4: Run extraction =====
pdf_path = "/content/សេចក្តីជូនដំណឹង_ស្តីពីការចេញចូលវិទ្យាសា្ថន_.pdf"
mixed_text = extract_text_from_pdf(pdf_path)

# ===== STEP 5: Output the text directly =====
print(mixed_text)
```
